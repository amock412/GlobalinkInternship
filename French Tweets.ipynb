# French Tweets
The summer project is related to social media comments surrounding urban infrastructure. The first exploration I did was just searching for certain key words on Twitter using the Twitter API. This is just playing around to see what data can be found via Twitter and how useful it may be in the future.

```python
import re
import tweepy
import pandas as pd
```

```python
consumer_key = '' # enter your information
consumer_secret = ''
access_token = ''
access_token_secret = ''
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth,wait_on_rate_limit=True)
print(api)
```

```python
def clean_tweet(tweet):
      # cleans tweet text by removing links and special characters
    return ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", " ", tweet).split())
```

```python
# search for tweets related to the word dirty in french 
tweets = tweepy.Cursor(api.search, q="sale", lang="fr").items(1000)


# box Montreal'-73.9278,45.3983,-73.2861,45.6247'
```

```python
# Store the tweets in a list
tweet_list = [tweet for tweet in tweets]
```

```python
# dataframe to save tweets
db_tweets = pd.DataFrame(columns = ['username', 'acctdesc', 'location', 'following',
                                        'followers', 'totaltweets', 'usercreatedts', 'tweetcreatedts',
                                        'retweetcount', 'text', 'hashtags'])
```

```python
for tweet in tweet_list[:500]:
# Pull the values
    username = tweet.user.screen_name
    acctdesc = tweet.user.description
    location = tweet.user.location
    following = tweet.user.friends_count
    followers = tweet.user.followers_count
    totaltweets = tweet.user.statuses_count
    usercreatedts = tweet.user.created_at
    tweetcreatedts = tweet.created_at
    retweetcount = tweet.retweet_count
    hashtags = tweet.entities['hashtags']         
    text = tweet.text
    
    ith_tweet = [username, acctdesc, location, following, followers, totaltweets,
                         usercreatedts, tweetcreatedts, retweetcount, text, hashtags]
    # Append to dataframe - db_tweets
    db_tweets.loc[len(db_tweets)] = ith_tweet
```
