As we can see, the Google Map reviews don't have exact dates for when they were written. However since the Yelp dataset provides exact dates, the dates have to be converted in a unique format that works for all of the data. Therefore all of the times in the Google Maps dataset are converted to approximate dates using relativedelta to compute what day ie 3 months would have been and use that as the date of a review. Of course this will not align with the real date of a review but it is close enough for our purpose. 

```python
import datetime
from dateutil.relativedelta import relativedelta
```

```python
pd.to_datetime(subwayGoogleMaps['date_retrieved'][0]) 
```

```python
relativedelta(months=-1) +pd.to_datetime(subwayGoogleMaps['date_retrieved'][0])
```

```python
def getDate(current,published):
    ''' converts a date like 3 months ago to a month,day, year format, it takes in the date of today (current) and the 
    date like 3 months ago to convert'''
    date_split = published.split()
    
    if date_split[1] == 'hours':
        numHours = int(date_split[0])
        timeDelta = datetime.timedelta(hours=numHours)
    
    elif date_split[1] == 'hour':
        timeDelta = datetime.timedelta(hours=1)
        
    elif date_split[1] == 'day':
        timeDelta = datetime.timedelta(days=1)
        
        
    elif date_split[1] == 'days':
        numDays = int(date_split[0])
        timeDelta = datetime.timedelta(days=numDays)
    elif date_split[1] == 'week':
        timeDelta = datetime.timedelta(weeks=1)
        
    elif date_split[1] == 'weeks':
        numWeeks = int(date_split[0])
        timeDelta = datetime.timedelta(weeks=numWeeks)
    elif date_split[1] == 'month':
        timeDelta = relativedelta(months=1)
    elif date_split[1] == 'months':
        numMonths = int(date_split[0])
        timeDelta = relativedelta(months=numMonths)
    elif date_split[1] == 'year':
        timeDelta = relativedelta(months=12)
    else: 
        #print(published)
        numYears = int(date_split[0])
        timeDelta = relativedelta(months=12*numYears)
        
    return current - timeDelta    
```

```python
# sample time conversion
getDate(pd.to_datetime(subwayGoogleMaps['date_retrieved'][0]),subwayGoogleMaps['published'][0])
```

```python
subwayReviews = subwayGoogleMaps.copy()
subwayReviews['date_retrieved'] = pd.to_datetime(subwayGoogleMaps['date_retrieved'])
```

```python
# review date conversion
reviewDate = subwayReviews.apply(lambda x: getDate(x['date_retrieved'],  x['published']),axis=1)
subwayReviews['review_date'] = reviewDate
```

```python
subwayReviews['type'] = 'Subway'
subwayReviews['scraped_from'] = 'Google Maps'
```

```python
parkReviews = parkGoogleMaps.copy()
parkReviews['date_retrieved'] = pd.to_datetime(parkGoogleMaps['date_retrieved'])
```

```python
reviewDateParks = parkReviews.apply(lambda x: getDate(x['date_retrieved'],  x['published']),axis=1)
```

```python
parkReviews['review_date'] = reviewDateParks 
parkReviews['type'] = 'Park'
parkReviews['scraped_from'] = 'Google Maps'
```

```python
# drop unnecessary columns
ubwayReviewsClean = subwayReviews.drop(columns=['review_id', 'published', 'date_retrieved', 'num_reviews'])
parkReviewsClean = parkReviews.drop(columns=['review_id', 'published', 'date_retrieved', 'num_reviews', 'label'])
```

```python
# merge google data 
mergedGoogleMaps = pd.concat([parkReviewsClean,subwayReviewsClean])
mergedGoogleMaps
```

```python
# add columns indicating language of reviews
parkYelpFr['lang'] = 'fr'

parkYelpEn['lang'] = 'en'
```

```python
# merging yelp dataset
mergedYelp = pd.concat([parkYelpFr, parkYelpEn])
mergedYelp
```
